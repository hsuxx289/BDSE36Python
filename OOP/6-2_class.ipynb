##### Class 設計

```python
import pymysql

class SqlQuery:

    def __init__(self, database=None):
        self.connection = pymysql.connect(
            host="localhost",
            user="",
            password= "",
            port=3306,
            database=database
        )

    def show_db(self):
        with self.connection.cursor() as cursor:
            sql = "SHOW DATABASES;"
            cursor.execute(sql)
            result = cursor.fetchall()
    
        return result

    def show_table(self):
        with self.connection.cursor() as cursor:
            sql = "SHOW TABLES;"
            cursor.execute(sql)
            result = cursor.fetchall()
    
        return result
    def sql_query(self, sql):
        with self.connection.cursor() as cursor:
            cursor.execute(sql)
            result = cursor.fetchall()
    
        return result


q1 = SqlQuery()
db = q1.show_db()
print(db)


sql = """
SELECT * FROM passengers
WHERE sex = 'male';

"""


my_train_titanic_q = SqlQuery("my_train_titanic")
print(my_train_titanic_q.sql_query(sql))

```
